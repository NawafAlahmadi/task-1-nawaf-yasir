# extract the pdf file

```python
import PyPDF2

with open(r'C:\Users\nawaf_alahmadi\Downloads\KSA_XYZ.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    text_ksa = ""
    for page in reader.pages:
        text_ksa += page.extract_text() 

print(text_ksa)
```

# tokenize the text 

```python
text_ksa_tok = word_tokenize(text_ksa)

print(text_ksa_tok)
```

# try ro make the text not list to work with the plot 

```python
text1 = ' '.join(text_ksa_tok)

text1
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS


wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = set (STOPWORDS),
                min_font_size = 10).generate(text1)

plt.figure(figsize = (8, 8), facecolor = 'red') 
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
